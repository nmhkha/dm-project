```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df = pd.read_csv("dirty_sample_500.csv")
data_df.head(10)
```

```python
data_df.columns
```

```python
data_df.isnull().sum()
```

```python
data_df['Air temperature [K]'].fillna(data_df['Air temperature [K]'].mean(), inplace=True)
data_df['Torque [Nm]'].fillna(data_df['Torque [Nm]'].mean(), inplace=True)
data_df.isnull().sum()
```

```python
data_df.head(10)
```

```python
data_df["Torque [Nm]"] = data_df["Torque [Nm]"].round(1)
data_df["Air temperature [K]"] = data_df["Air temperature [K]"].round(1)
data_df.head(10)
```

```python
# Liệt kê tất cả giá trị khác nhau trong cột Type 
data_df['Type'].unique()
```

```python
# thống kê số lượng từng giá trị
data_df['Type'].value_counts()
```

```python
# Chuẩn hóa chữ hoa/thường
data_df['Type'] = data_df['Type'].str.upper()
```

```python
data_df['Type'].unique()
```

```python
# Xóa khoảng trắng thừa
data_df['Type'] = data_df['Type'].str.strip()
```

```python
data_df['Type'].unique()
```

```python
# Chuyển giá trị lỗi thành NAN 
data_df['Type'] = data_df['Type'].replace('??', np.nan)
```

```python
df1 = data_df.dropna()
df1.isnull().sum()
```

```python
# Giữ lại giá trị hợp lệ
valid_types = ['L', 'M', 'H']
df1 = df1[df1['Type'].isin(valid_types) | df1['Type'].isna()]
```

```python
df1['Type'].unique()
```

```python
# Các giá trị outlier
mean = df1['Air temperature [K]'].mean()
std = df1['Air temperature [K]'].std()
outliers = df1[(df1['Air temperature [K]'] > mean + 3*std) |
(df1['Air temperature [K]'] < mean - 3*std)]
outliers
```

```python
df1 = df1[df1['Air temperature [K]'] <= 400]
```

```python
df1.to_csv("cleaned_data.csv", index=False, encoding='utf-8-sig')
```
