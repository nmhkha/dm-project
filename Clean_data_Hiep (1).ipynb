```python
df1 = data_df.dropna()
df1.isnull().sum()
```

```python
# Giữ lại giá trị hợp lệ
valid_types = ['L', 'M', 'H']
df1 = df1[df1['Type'].isin(valid_types) | df1['Type'].isna()]
```

```python
cols = ["Air temperature [K]","Rotational speed [rpm]", "Process temperature [K]", "Torque [Nm]", "Tool wear [min]"]

for col in cols:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df1[(df1[col] < lower_bound) | (df1[col] > upper_bound)]
    print(f"{col}: {len(outliers)} giá trị ngoại lai")
```

```python
from sklearn.ensemble import RandomForestRegressor

def replace_outliers_with_prediction(df1, col, features):
    # 1. Tính Q1 và Q3
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1

    # 2. Xác định ngưỡng dưới và ngưỡng trên để phát hiện outlier
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # 3. Tạo mask (True/False) cho các hàng là outlier
    mask_outlier = (df1[col] < lower) | (df1[col] > upper)
    
    # 4. Chia dữ liệu thành 2 phần: train_data (không outlier) và test_data (có outlier)
    train_data = df1[~mask_outlier]
    test_data = df1[mask_outlier]
    
    # 5. Nếu có outlier thì mới tiến hành dự đoán
    if not test_data.empty:
        # 5.1. Khởi tạo mô hình Random Forest Regressor
        model = RandomForestRegressor()

        # 5.2. Huấn luyện mô hình dựa trên các cột features để dự đoán giá trị col
        model.fit(train_data[features], train_data[col])

        # 5.3. Dự đoán giá trị mới cho các outlier
        predicted_values = model.predict(test_data[features])

        # 5.4. Gán giá trị dự đoán vào đúng vị trí của outlier trong DataFrame
        df1[col] = df1[col].astype(float)
        df1.loc[mask_outlier, col] = predicted_values


# Danh sách các cột có outlier
cols_outlier = ["Air temperature [K]", "Rotational speed [rpm]", "Torque [Nm]"]

# Các cột dùng để dự đoán
features = ["Process temperature [K]", "Tool wear [min]"]

# Chạy thay thế outlier cho từng cột cho đến khi hết
for col in cols_outlier:
    while True:
        Q1 = df1[col].quantile(0.25)
        Q3 = df1[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        mask_outlier = (df1[col] < lower) | (df1[col] > upper)

        if mask_outlier.sum() == 0:
            print(f"{col}: không còn outliers")
            break
        
        replace_outliers_with_prediction(df1, col, features)
```

```python
df1["Air temperature [K]"] = df1["Air temperature [K]"].round(1)
df1["Process temperature [K]"] = df1["Process temperature [K]"].round(1)
df1["Rotational speed [rpm]"] = df1["Rotational speed [rpm]"].round(1)
df1["Torque [Nm]"] = df1["Torque [Nm]"].round(1)
df1["Tool wear [min]"] = df1["Tool wear [min]"].round(1)
df1.head(10)
```
